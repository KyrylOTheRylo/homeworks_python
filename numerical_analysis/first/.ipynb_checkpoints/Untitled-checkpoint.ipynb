```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
A = 0
B = 5
N = 100
EPS = 1.7*10e-2
def f(x):
    tmp = np.random.normal(0, EPS)
    return  2*np.sin(2*x) + tmp *3
```

```python
x = np.linspace(A,B,N)
x
```

```python
y = list(map(f, x))
y
```

```python
plt.scatter(x,y)
```

```python
def polyfit_custom(x, y, deg):
    # Ensure x and y have the same length
    if len(x) != len(y):
        raise ValueError("Input arrays x and y must have the same length")

    # Create the Vandermonde matrix
    A = np.vander(x, deg + 1)

    # Solve the linear system of equations
    coefficients = np.linalg.lstsq(A, y, rcond=None)[0]

    return coefficients
```

```python
pf = polyfit_custom(x,y,7)
pf
```

```python
def poly_from_fit(numbers,x):
    t =  x**(len(numbers) - 1)

    if x == 0:
        return numbers[-1]
    if len(numbers)>0:
        answer = []
        for i in range(len(numbers)):
            
            answer.append(t*numbers[i])
            t /= x
        return np.sum(answer)
    else:
        return numbers
            
```

```python
answer_poly = [poly_from_fit(pf,i) for i in x]
```

```python
plt.plot(x, answer_poly, c="r")
plt.scatter(x,y)
plt.show()
```
